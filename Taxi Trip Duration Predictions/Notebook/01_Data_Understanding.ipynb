# Data Understanding

## Dataset Information

Dataset yang digunakan adalah NYC Taxi Trip Data 2025 Sample yang berisi informasi historis perjalanan taksi di New York City.

Tujuan analisis adalah membangun model Machine Learning untuk memprediksi durasi perjalanan (trip duration) guna membantu perusahaan transportasi meningkatkan efisiensi operasional.

## Business Problem

Perusahaan transportasi mengalami kesulitan memperkirakan durasi perjalanan secara akurat sehingga mengganggu perencanaan operasional.

## Machine Learning Problem

Membangun model regresi untuk memprediksi durasi perjalanan berdasarkan karakteristik perjalanan pelanggan.

## Dataset Summary

| Informasi | Hasil |
|-----------|-------|
| Jumlah data | 3.946.698 baris |
| Jumlah fitur | 25 kolom |
| Missing value | 757.387 data |
| Kolom yang memiliki missing value | Airport_fee |
| Duplicate data | 0 |
| Target variabel | duration_min |
| Jenis permasalahan | Regression |

## Temuan Awal

1. Dataset memiliki **3.946.698 data perjalanan** sehingga cukup besar untuk membangun model prediksi yang stabil.

2. Dataset memiliki **25 fitur** yang terdiri dari:

- 13 kolom numerik (`float64`)
- 7 kolom numerik (`int64`)
- 4 kolom kategorikal (`object/string`)
- 1 kolom boolean

3. Tidak ditemukan data duplikat.

4. Terdapat missing value pada kolom `Airport_fee` sebanyak **757.387 data**.

5. Target prediksi yang akan digunakan adalah `duration_min`.

6. Proyek ini termasuk ke dalam **Supervised Machine Learning - Regression Problem**.
